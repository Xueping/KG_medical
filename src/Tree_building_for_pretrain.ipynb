```python
import pickle
import torch
import time
import random
import numpy as np
import torch.optim as optim
from GRAM.gram_helpers import train_model, calculate_dimSize, get_rootCode, build_tree, load_data, print2file, build_tree_with_padding
from GRAM.gram_module import GRAM as gram_model
import torch.nn as nn
from GRAM.gram_module import DAGAttention, DAGAttention2D
```

```python
dir_path = '../'
tree_file = dir_path + 'outputs/kemce/data/mimic/mimic_pre_train'
dict_file = dir_path + 'outputs/kemce/data/mimic/mimic_pre_train.types'
```

```python
embd_dim_size = 100
attn_dim_size = 100
rnn_dim_size = 100
# Batch size for training (change depending on how much memory you have)
batch_size = 64
# Number of epochs to utils for
num_epochs = 100
seed = 42
```

```python
mask_list = []
leaves_list = []
ancestors_list = []
for i in range(5, 0, -1):
    leaves, ancestors, masks = build_tree_with_padding(tree_file + '.level' + str(i) + '.pk')
    leaves_list.extend(leaves)
    ancestors_list.extend(ancestors)
    mask_list.extend(masks)
```

```python
num_tree_nodes = get_rootCode(tree_file+'.level2.pk') + 1
```

```python
num_tree_nodes
```

```python
# embedding for leaf codes and internal codes
embed_init = nn.Embedding(num_tree_nodes, 100)
# DAG attention
dag_attention = DAGAttention2D(2*embd_dim_size, attn_dim_size)
```

```python
leaves_emb = embed_init(torch.tensor(leaves_list).long())
ancestors_emb = embed_init(torch.tensor(ancestors_list).long())
mask_tensor = torch.tensor(mask_list).float()
```

```python
dag = dag_attention(leaves_emb, ancestors_emb, mask_tensor)
```
